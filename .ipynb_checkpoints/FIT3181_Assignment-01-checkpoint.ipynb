####  <span style="color:red">**Question 1.1**</span> **Activation function plays an important role in modern Deep NNs. For each of the activation function below, state its output range and find its derivative (show your steps)**

<span style="color:red">**(a)**</span> Sigmoid: $\sigma(x) = \frac{1}{1+\text{exp}{(-x)}}$ 
<div style="text-align: right"><span style="color:red">[1.5 points]</span></div> 

<b><font color="blue" weight="bold">Answer</font></b>

Output range: (0, 1)

$\sigma(x) = \frac{1}{1+\text{exp}{(-x)}} = (1+\text{exp}{(-x)})^{-1}$

$\sigma$'(x)

= $-(1+\text{exp}{(-x)})^{-2}(-\text{exp}{(-x)})$

= $(1+\text{exp}{(-x)})^{-2}(\text{exp}{(-x)})$

= $\frac{\text{exp}{(-x)}}{(1+\text{exp}{(-x)})^{2}}$

<span style="color:red">**(b)**</span> Tanh: $\sigma(x) = \frac{\exp(x) - \exp{(-x)}}{\exp(x) + \exp{(-x)}}$
<div style="text-align: right"><span style="color:red">[1.5 points]</span></div> 

<b><font color="blue" weight="bold">Answer</font></b>

Output range: (-1, 1)

$\sigma$'(x)

= $\frac{(\exp(x) + \exp{(-x)})(\exp(x) - (-1)\exp{(-x)}) - (\exp(x) - \exp{(-x)})(\exp(x) + (-1)\exp{(-x)})}{(\exp(x) + \exp{(-x)})^{2}}$

= $\frac{(\exp(x) + \exp{(-x)})(\exp(x) + \exp{(-x)}) - (\exp(x) - \exp{(-x)})(\exp(x) - \exp{(-x)})}{(\exp(x) + \exp{(-x)})^{2}}$ 

= $\frac{(\exp(x) + \exp{(-x)})^{2} - (\exp(x) - \exp{(-x)})^{2}}{(\exp(x) + \exp{(-x)})^{2}}$ 

= $\frac{(\exp(2x) + 2\exp(x)\exp(-x) +\exp{(-2x)}) - (\exp(2x) - 2\exp(x)\exp(-x) + \exp{(-2x)})}{(\exp(x) + \exp{(-x)})^{2}}$

= $\frac{4\exp(x)\exp(-x)}{(\exp(x) + \exp{(-x)})^{2}}$ 

= $\frac{4}{(\exp(x) + \exp{(-x)})^{2}}$

<font size=2> 
<b><font color="blue" weight="bold">Answer</font></b>
  
Input: ($x$, $y$)

$\hat{y} = forward(x)$

$\hat{y}_k = w^o_{k1}h_1 + w^o_{k2}h_2$

Goal: minimize $J_t(w) = \frac{1}{2} \sum_{k=1}^{K}{(\hat{y}_k - y_k)^{2}}$ <br>

***
#### <font color="green">For $w^o_{11}:$</font>

We know that...

<b>(1)</b> $\hat{y}_1 = w^o_{11}h_1 + w^o_{12}h_2$

<b>(2)</b> $\begin{equation*} 
\frac{\partial \hat{y}_1}{\partial w^o_{11}} \\
= h_1 
\end{equation*}$

<b>(3)</b> $\begin{equation*}
\frac{\partial J_t(w)}{\partial \hat{y}_1} \\
=\frac{\partial J_t(w)_1}{\partial \hat{y}_1} + \frac{\partial J_t(w)_2}{\partial \hat{y}_1} + \ldots + \frac{\partial J_t(w)_K}{\partial \hat{y}_1}\\
= \frac{\partial \left(\frac{1}{2} (\hat{y}_1 - y_1)^{2} \right)}{\partial \hat{y}_1} + 0 + \ldots + 0 \\
= (\hat{y}_1 - y_1)
\end{equation*}$

By using the fact above, we can conclude that...

\begin{align*}
\frac{\partial J_t(w)}{\partial w^o_{11}} 
&= \frac{\partial J_t(w)}{\partial \hat{y}_1} \times \frac{\partial \hat{y}_1}{\partial w^o_{11}} \\
&= (\hat{y}_1 - y_1)h_1
\end{align*}

$\therefore$ `GD update for` $w^o_{11} \leftarrow w^o_{11} - \eta(\hat{y}_1 - y_1)h_1$

***
#### <font color="green">For $w^o_{k1}:$</font>

We know that...

`GD update for` $w^o_{11} \leftarrow w^o_{11} - \eta(\hat{y}_1 - y_1)h_1$

`GD update for` $w^o_{21} \leftarrow w^o_{21} - \eta(\hat{y}_2 - y_2)h_1$

&ensp; $\vdots$ &ensp; (Same goes to $w^o_{31}, w^o_{41}, \ldots , w^o_{K1}$)

Therefore, for $w^o_{k1}:$

<span style="color:red">**(c)**</span> What are the gradient descent (GD) update equation for $w^h_{11}$ and $w^h_{1m}$ in general for $m=1, \ldots, M$ (show your derivation)? 
<div style="text-align: right"><span style="color:red">[2.5 points]</span></div>

<font size=2>
<b><font color="blue" weight="bold">Answer</font></b>

Input: ($x$, $y$)

$\hat{y} = forward(x)$

$\hat{y}_k = w^o_{k1}h_1 + w^o_{k2}h_2$

$h_1 = \sigma(\bar{h_1})$ where $\bar{h_1} = \sum_{m=1}^{M}{(w^h_{1m} \times x_m)}$

Goal: minimize $J_t(w) = \frac{1}{2} \sum_{k=1}^{K}{(\hat{y}_k - y_k)^{2}}$

#### <font color="green">For $w^h_{11}:$</font> 

We know that...

<b>(1)</b> $\begin{equation*} 
\frac{\partial \bar{h_1}}{\partial w^h_{11}} \\
= \frac{\partial \left(w^h_{11} \:\times\: x_1 \:+\: w^h_{12} \:\times\: x_2 \:+\: \ldots \:+\: w^h_{1M} \:\times\: x_M \right)}{\partial w^h_{11}} \\
= x_1
\end{equation*}$

<b>(2)</b> $\begin{equation*} 
\frac{\partial h_1}{\partial \bar{h_1}} \\
= \frac{\partial \sigma(\bar{h_1})}{\partial \bar{h_1}} \\ 
= \sigma(\bar{h_1})(1-\sigma(\bar{h_1})) \\ 
= h_1(1-h_1)
\end{equation*}$

<b>(3)</b> $\begin{equation*} 
\frac{\partial \hat{y}_k}{\partial h_1} \\
= w^o_{k1}
\end{equation*}$

<b>(4)</b> $\begin{equation*} 
\frac{\partial J_t(w)}{\partial \hat{y}_k} \\
= (\hat{y}_k - y_k)
\end{equation*}$ <font color="purple">(derived from 1.3b equation (3))</font>

<b>(5)</b> $\begin{equation*} 
\frac{\partial J_t(w)_k}{\partial h_1} \\
= \frac{\partial J_t(w)_k}{\partial \hat{y}_k} \times \frac{\partial \hat{y}_k}{\partial h_1} \\
= (\hat{y}_k - y_k)(w^o_{k1})
\end{equation*}$

<b>(6)</b> $\begin{equation*}
\frac{\partial J_t(w)}{\partial h_1} \\
= \frac{\partial J_t(w)_1}{\partial h_1} + \frac{\partial J_t(w)_2}{\partial h_1} + \ldots + \frac{\partial J_t(w)_K}{\partial h_1} \\
= (\hat{y}_1 - y_1)(w^o_{11}) + (\hat{y}_2 - y_2)(w^o_{21}) + \ldots + (\hat{y}_K - y_K)(w^o_{K1}) \\
= \begin{bmatrix}
\hat{y}_1 - y_1 \\
\hat{y}_2 - y_2 \\
\vdots \\
\hat{y}_K - y_K \\
\end{bmatrix} 
\cdot
\begin{bmatrix}  
w^o_{11} \\
w^o_{21} \\
\vdots \\
w^o_{K1}
\end{bmatrix} \\
= \left(\begin{bmatrix}
\hat{y}_1 \\
\hat{y}_2 \\
\vdots \\
\hat{y}_K \\
\end{bmatrix} -
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_K \\
\end{bmatrix}
\right)
\cdot
\begin{bmatrix}  
w^o_{11} \\
w^o_{21} \\
\vdots \\
w^o_{K1}
\end{bmatrix} \\
= (\hat{y} - y)\cdot(w^o_{1})
\end{equation*}$ <font color="purple">($w^o_{1}$ = a vector that contains all the weights from $h_1$)</font>

By using the fact above, we can conclude that...

\begin{align*}
\frac{\partial J_t(w)}{\partial w^h_{11}} 
&= \frac{\partial J_t(w)}{\partial \bar{h_1}} \times \frac{\partial \bar{h_1}}{\partial w^h_{11}} \\
&= \frac{\partial J_t(w)}{\partial h_1} \times \frac{\partial h_1}{\partial \bar{h_1}} \times \frac{\partial \bar{h_1}}{\partial w^h_{11}} \\
&= [(\hat{y} - y)\cdot(w^o_{1})](h_1)(1-h_1)(x_1)
\end{align*}

$\therefore$ `GD update for` $w^h_{11} \leftarrow w^h_{11} - \eta[(\hat{y} - y)\cdot(w^o_{1})](h_1)(1-h_1)(x_1)$

`GD update for` $w^h_{11} \leftarrow w^h_{11} - \eta[(\hat{y} - y)\cdot(w^o_{1})](h_1)(1-h_1)(x_1)$

&ensp; $\vdots$ &ensp; (Same goes to $w^h_{13}, w^h_{14}, \ldots , w^h_{1M}$)

Therefore, for $w^h_{1m}:$ <br>

$\therefore$ `GD update for` $w^h_{1m} \leftarrow w^h_{1m} - \eta[(\hat{y} - y)\cdot(w^o_{1})](h_1)(1-h_1)(x_m)$

**For question 3.3 to 3.6, you'll need to write your own model in a way that makes it easy for you to experiment with different architectures as well as parameters. The goal is to be able to pass the parameters to initialize a new instance of `YourModel` to build different network architectures with different parameters. Below is descriptions of some parameters for `YourModel`, which you can find in function `__init__()` for the class `DefaultModel`:**

1. `num_blocks`: an integer specifying the number of blocks in our network. Each block has the pattern `[conv, batch norm, activation, conv, batch norm, mean pool, dropout]`. All convolutional layers have filter size $(3, 3)$, strides $(1, 1)$ and 'SAME' padding, and all mean pool layers have strides $(1, 2)$ and 'SAME' padding. The network will consists of a few blocks before applying a linear layer to output the logits for the softmax layer.

2. `feature_maps`: the number of feature maps in the first block of the network. The number of feature_maps will double in each of the following block. To make it convenient for you, we already calculated the number of feature maps for each block for you in line $162$
3. `keep_prob`: the keep probability for dropout. Setting `keep_prob` to $1.0$ means not using dropout. Notice that dropout is not used in the inference phase. Therefore, you should pass the placeholder `self.keep_prob_holder` (line $203$) to the dropout layers. The place holder will take the value of `keep_prob` during training phase (line $270$) and the value of $1.0$ during inference phase (line $326$ and $332$).
4. `batch_norm`: the batch normalization function to be use, such as tf.layers.batch_normalization. Setting `batch_norm` to `None` means not using batch normalization. Notice that the batch normalization should take a parameter `training`, which specifies whether the computation is in the training (True) or inference (False) phase, and which is passed to the computational graph through `self.phase_train` (see line $204$, $270$, $327$) 

**You'll only need to re-write the code for the `build_cnn` function for the `YourModel` class from the cell below.**

```python
# YOU ARE REQUIRED TO INSERT YOUR CODES IN THIS CELL

import tensorflow as tf

class YourModel(DefaultModel):
    def build_cnn(self, x, reuse=False, name='CNN'):
        # ENTER YOUR CODES HERE 
        with tf.variable_scope(name, reuse=reuse):

            # for each block            
            for i in range(self.num_blocks):

                h = Layers.conv2D(x, self.feature_maps[i], kernel_size=3, strides=1, name='block'+str(i)+'.conv0')
                tf.summary.histogram('block'+str(i)+'.conv0', h) 
                if self.batch_norm is not None:
                    h = self.batch_norm(h, self.phase_train, name='block'+str(i)+'.bn0')
                    tf.summary.histogram('block'+str(i)+'.bn0', h)
                h = self.activation_func(h, name='block'+str(i)+'.relu0')
                tf.summary.histogram('block'+str(i)+'.relu0', h) 

                h = Layers.conv2D(h, self.feature_maps[i], kernel_size=3, strides=1, name='block'+str(i)+'.conv1')
                tf.summary.histogram('block'+str(i)+'.conv1', h) 
                if self.batch_norm is not None:
                    h = self.batch_norm(h, self.phase_train, name='block'+str(i)+'.bn1')
                    tf.summary.histogram('block'+str(i)+'.bn1', h)
                h = self.activation_func(h, name='block'+str(i)+'relu1')
                tf.summary.histogram('block'+str(i)+'.relu1', h)
                h = Layers.mean_pool(h, name='block'+str(i)+'.avg_pool')
                h = Layers.dropout(h, self.keep_prob_holder, name='block'+str(i)+'.dropout')           
                
                x=h
                
            # after all block
            # reshape and use a fully connected layer
            h = tf.reshape(h, [-1, h.get_shape()[1] * h.get_shape()[2] * h.get_shape()[3]])

            # now output the logits for softmax classification
            logits = Layers.dense(h, self.num_classes, name='output.logits')
            tf.summary.histogram('output.logits', logits)
            return logits
        
```
